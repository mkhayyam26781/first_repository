```python
import pandas as pd
data = pd.read_excel('donor.xlsx')
print(data.head())
```

```python
data['score'] = (
    0.4*(1/data['distance'])+
    0.3*(data['availability'])+
    0.3*(data['last_donation_days']/100)
)
best = data.sort_values(by='score', ascending=False)

print("THE BEST DONORS:")
print(best.head(1)['name'])
```

```python
filtered = data[
    (data['city'] == "Peshawar") &
    (data['blood_group'] == "O+")
]

best = filtered.sort_values(by='score', ascending=False)

print(best.head(1))
```

```python
city = input("Enter City: ")
blood = input("Enter Blood Group: ")

filtered = data[
    (data['city'] == city) &
    (data['blood_group'] == blood)
]

best = filtered.sort_values(by='score', ascending=False)

print(best.head(1))
```

```python
best_donor = best.iloc[0]

print("Best Donor Found:")
print("Name:", best_donor['name'])
print("City:", best_donor['city'])
print("Blood Group:", best_donor['blood_group'])
```

```python
best.to_excel("best_donor.xlsx")
```
